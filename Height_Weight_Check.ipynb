# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
m = reg.coef_ #y= mx+c
```

```python
m
```

```python
c = reg.intercept_
c
```

```python
#y= mx+c
```

```python
y = m*2100 + c
y
```

```python
df["Predicted Result"] = reg.predict(x)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```
